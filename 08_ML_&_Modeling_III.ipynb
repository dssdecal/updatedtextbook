# ML & Modeling III: Deep Learning and Representations

### 1. From Features to Representations

In traditional machine learning, models rely heavily on feature engineering. Practitioners manually design input features that they believe will be useful for prediction. For example, when predicting housing prices, one might include features such as square footage, number of bedrooms, and location. The model then learns how to combine these features to make predictions.

Deep learning fundamentally changes this paradigm.

Instead of requiring humans to design features, deep learning models learn useful representations directly from raw data. These representations are learned automatically through training.

For example:

*   In image tasks, models learn to transform raw pixel values into edges, shapes, and eventually full objects.
*   In text tasks, models transform words into representations that capture meaning and context.

This shift from manually engineered features to learned representations is one of the defining characteristics of modern machine learning.

### 2. Neural Networks and Learning

At the core of deep learning are neural networks. A neural network is a model composed of multiple layers, where each layer transforms the input data into a new representation.

Each layer applies a transformation to its input, and as data moves through the network, it becomes increasingly structured and useful for the task at hand. Early layers often capture simple patterns, while deeper layers capture more complex and abstract patterns.

Neural networks learn through an iterative process:


*   Forward pass: The input is passed through the network to produce a prediction.
*   Loss function: The prediction is compared to the true value, producing an error.

*   Backward pass (backpropagation): The model determines how each parameter contributed to the error.
*   Parameter update: The model adjusts its parameters to reduce the error.

This cycle is repeated many times over the dataset. Over time, the network improves its ability to produce accurate predictions.

### 3. Intuition Behind Backpropagation

Backpropagation is the mechanism that allows neural networks to learn.

When a model makes a prediction, it is rarely perfect. Backpropagation provides a way to distribute the error across the network so that each parameter knows how it should change.

The key idea is to assign responsibility for the error:


*   If a parameter contributed heavily to the mistake, it should be adjusted more.
*   If it contributed only slightly, it should be adjusted less.

By repeatedly assigning responsibility and updating parameters, the network gradually improves.

Importantly, these updates are typically small. Learning happens through many small adjustments rather than large, abrupt changes. This makes the training process stable and allows the model to converge to a good solution.

### 4. Attention and Transformers

Early neural network architectures for sequential data, such as recurrent neural networks (RNNs), processed inputs one step at a time. This made it difficult to capture long-range dependencies and limited the model’s ability to understand complex relationships in data.

Transformers introduced a new mechanism called attention.

Attention allows a model to consider all parts of the input simultaneously and determine which parts are most relevant. Instead of processing data sequentially, the model can directly learn relationships between any two elements in the input.

For example, in the sentence:
“The animal didn’t cross the street because it was tired,”
the word “it” refers to “the animal.” Attention mechanisms allow the model to identify this relationship, even though the relevant words are separated.

Transformers are built using layers of attention mechanisms combined with standard neural network components. This architecture has become the foundation of many modern AI systems, particularly in natural language processing.

### 5. Embeddings and Representation Learning

A central concept in modern machine learning is the idea of an embedding.

An embedding is a numerical vector that represents data in a way that captures its meaning. These vectors allow models to process complex data such as text and images.

In natural language processing:


*   Words are mapped to vectors
*   Words with similar meanings are placed close together in this vector space


For example, the words “king” and “queen” would have similar representations, while unrelated words would be farther apart.

In computer vision:


*   Images are transformed into vectors that capture visual features
*   Early representations may capture edges or textures
*   Later representations capture objects or semantic meaning

The key idea is that all types of data (text, images, and more) can be converted into embeddings. Once in this form, machine learning models can operate on them in a unified way.

### 6. A Unified View of Modern AI

Modern machine learning systems can be understood through a common pipeline:


*   Raw data (text, images, etc.)
*   Converted into embeddings
*   Processed by a neural network
*   Produces an output (prediction, classification, or generation)

This framework applies across many domains, including:

*   Language models
*   Image recognition systems
*   Recommendation systems

Despite differences in application, these systems share the same underlying principles: representation learning, neural networks, and optimization through backpropagation.

### 7. Summary

Deep learning represents a shift from manually designing features to automatically learning representations from data.

Neural networks enable this process by transforming data through multiple layers, while backpropagation allows the model to learn by assigning responsibility for errors.

Transformers and attention mechanisms further enhance this capability by allowing models to capture complex relationships within data.

Finally, embeddings provide a unified way to represent different types of data, forming the foundation of modern AI systems.
